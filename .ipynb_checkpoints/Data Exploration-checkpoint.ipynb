# Data Exploration

```python
# Import statements
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

First, we read in the csv file that we created from the data collection and cleaning notebook.

```python
final_df = pd.read_csv("/Users/alickzhang/desktop/dataproject/nba_player_ml/player.csv")
final_df.head()
```

### How are position split by team?

```python
pos_counts = final_df["pos"].value_counts()
pos_counts.plot.pie()
```

**Conclusion:**

The positions are divided evenly for all the players in the NBA.

### What are the positions that have the highest average points per game?

```python
pos_pts_cube = final_df.pivot_table(
    index="pos",
    values="pts_per_g", aggfunc=np.mean)
pos_pts_cube.sort_values("pts_per_g").plot.bar(figsize=(20,8))
```

**Conclusion:**

The shooting guard and point guard tend to have the highest average points per game.

### What is the relationship between season, positions, and points per game?

```python
season_pts_cube = final_df.pivot_table(
    index="season", columns="pos",
    values="pts_per_g", aggfunc=np.mean)
season_pts_cube.plot.bar(figsize=(20,8))
```

### Explore the trend of data for variables related to points
- FGA Per Game by Season

```python
team_div_df = pd.read_csv("team_div.csv")

team_div_df.groupby("season")["fga_per_g"].mean().plot.line()
fig = plt.gcf()
fig.set_size_inches(16,6)
ax = plt.gca()
ax.set_xlabel("Season")
ax.set_ylabel("Average FGA Per Game")
```

- Points per Game by Season

```python
team_div_df.groupby("season")["pts_per_g"].mean().plot.line()
fig = plt.gcf()
fig.set_size_inches(16,6)
ax = plt.gca()
ax.set_xlabel("Season")
ax.set_ylabel("Average FGA Per Game")
```

### What caused this increase in points/pace?

- 3 Point Attemped by Conference and Season
    - We see a positive increase in 3 point field goals attemped across all divisions over the seasons

```python
team_div_df.groupby(["season", "division"])["fg3a_per_g"].mean().unstack().plot.bar()
fig = plt.gcf()
fig.set_size_inches(16,6)
plt.xlabel("")
plt.ylabel("3 Points Attempted")
```

### What teams are utilizing this new style of basketball?
- HOU seems to make the most 3 point shots per game compared to all other teams

```python
team_div_df.boxplot("fg3_per_g", by = "team_id")
fig = plt.gcf()
fig.set_size_inches(16,6)
ax = plt.gca()
plt.title("")
ax.set_ylabel("3 Pointers Made")
ax.set_xlabel("")
```

### Which team is most efficient at shooting 3's?
- Warriors seem to have the highest 3 point shooting percentage. Surprising to see Houston not be in the top 10.


```python
team_div_df.boxplot("fg3_pct", by = "team_id")
fig = plt.gcf()
fig.set_size_inches(16,6)
ax = plt.gca()
plt.titlae("")
ax.set_ylabel("3 Point %")
ax.set_xlabel("")
```
